### Dataset Overview

The Superstore sales dataset contains detailed information about sales transactions at a retail store. The data spans several years and includes various attributes related to the sales process, customer details, product categories, and regional information. This dataset is crucial for analyzing sales trends, identifying best-selling products, and understanding regional performance.

#### Attributes:

- **order_id**: A unique identifier for each order.
- **order_date**: The date the order was placed.
- **ship_date**: The date the order was shipped.
- **ship_mode**: The mode of shipment (e.g., Standard Class, Second Class).
- **customer_name**: The name of the customer who placed the order.
- **segment**: The market segment of the customer (e.g., Consumer, Corporate, Home Office).
- **state**: The state where the order was placed.
- **country**: The country where the order was placed.
- **market**: The market region (e.g., APAC, EMEA).
- **region**: The geographical region (e.g., North, West).
- **category**: The category of the product sold (e.g., Office Supplies, Furniture).
- **sub_category**: The sub-category of the product sold (e.g., Storage, Binders).
- **product_name**: The name of the product sold.
- **sales**: The sales amount for the transaction.
- **quantity**: The quantity of products sold.
- **discount**: The discount applied on the sale.
- **profit**: The profit made on the sale.
- **shipping_cost**: The cost of shipping for the transaction.
- **order_priority**: The priority level of the order (e.g., High, Medium).
- **year**: The year in which the order was placed.

# Item 2: Sales Trends Over Time, Best Selling Products, and Regional Performance.

````python
# Inspect the column names to find the correct name for the order date column
print(sales_df.columns)
````

````python

### Full Code in One Place

# Here is the full code to run all three analyses sequentially in one notebook cell:

# ```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (assuming the file is in the data directory)
data_path = '../data/superstore_sales.csv'
sales_df = pd.read_csv(data_path)

# Convert the order date to datetime format using the correct column name
sales_df['order_date'] = pd.to_datetime(sales_df['order_date'])

# Ensure the sales column is numeric
sales_df['sales'] = pd.to_numeric(sales_df['sales'], errors='coerce')

# Set the order date as the index
sales_df.set_index('order_date', inplace=True)

# Sales Trends Over Time
monthly_sales = sales_df['sales'].resample('M').sum()
plt.figure(figsize=(12, 6))
monthly_sales.plot()
plt.title('Sales Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
print("Summary Statistics for Sales Over Time:")
print(monthly_sales.describe())

# Best Selling Products
product_sales = sales_df.groupby('product_name')['sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
product_sales.head(10).plot(kind='bar')
plt.title('Top 10 Best Selling Products')
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
print("Top 10 Best Selling Products:")
print(product_sales.head(10))

# Regional Performance Analysis
region_sales = sales_df.groupby('region')['sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
region_sales.plot(kind='bar')
plt.title('Regional Performance')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
print("Regional Performance:")
print(region_sales)
````
